# Satchel Example

This notebook walks you through how to use the Satchel package and it's various features. The model is still being updated so expect things to change in the future.

```python
from satchel.model import Satchel
from satchel import utils, SatchelComparison
```

```python
# set up model run
model = Satchel(seed = 759, use_current_results=False)  # set a seed for reproduceability
results = model.simulate()
```

## Results

The `SatchelResults` class that gets returned from a model run has a variety of built in plotting methods as well as a summary of the season results that you can break down by division

```python
results.boxplot()
```

```python
results.results_grid();
```

```python
results.results_scatter('HOU', offset=1000)
```

Use the `season_percentile` method to see what percentile of the wins distribution a team's season will fall.

```python
results.season_percentile(team='HOU', wins=90)
```

## Test the effects of a transaction

Satchel has the ability to test what would happen if there were a trade. As an example, suppose the Cubs traded Dansby Swanson to the Dodgers for Mookie Betts at the trade deadline (yes, I know this is unrealistic).

First, look up the IDs for each player.

```python
print(utils.player_id_lookup('swanson', 'dansby'))
print(utils.player_id_lookup('betts', 'mookie'))
```

Then set up the transaction dictionary and pass it into Satchel

```python
transaction = {'18314': {'team': 'LAD', 'date': '2024-07-30'},
               '13611': {'team': 'CHC', 'date': '2024-07-30'},}
model2 = Satchel(seed = 759, use_current_results=False, transactions=transaction)
results2 = model2.simulate()
```

Then use the `SatchelComparison` class to compare the results.

```python
comp = SatchelComparison(results, results2)
```

```python
comp.win_dist_chart(team='CHC', sim1_label='Base', sim2_label='Trade');
```

```python
comp.win_dist_chart(team='LAD', sim1_label='Base', sim2_label='Trade');
```
